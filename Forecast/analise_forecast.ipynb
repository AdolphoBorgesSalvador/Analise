```python
import pandas as pd

# Supondo que forecast_df seja o seu DataFrame inicial com as colunas 'Forecast', 'Saidas', 'Canal', etc.

# Convertendo as colunas para numérico e 'Data FCST' para datetime
forecast_df['Forecast'] = pd.to_numeric(forecast_df['Forecast'], errors='coerce')
forecast_df['Saidas'] = pd.to_numeric(forecast_df['Saidas'], errors='coerce')
forecast_df['Data FCST'] = pd.to_datetime(forecast_df['Data FCST'], errors='coerce')

# Filtrando para valores de 'Saidas' maiores que zero
forecast_df_filtered = forecast_df[forecast_df['Saidas'] > 0].copy()

# Calculando o Erro Quadrático Médio (MSE) e o RMSE para cada canal
def calculate_mse_rmse(group):
    # Calculando o erro quadrático para cada linha
    group['Erro Quadrático'] = (group['Forecast'] - group['Saidas']) ** 2
    # Calculando o MSE (média do erro quadrático)
    mse = group['Erro Quadrático'].mean()
    # Calculando o RMSE como a raiz quadrada do MSE
    rmse = mse ** 0.5
    return pd.Series({'MSE': mse, 'RMSE': rmse})

# Aplicando a função por Canal
mse_rmse_by_canal = forecast_df_filtered.groupby('Canal').apply(calculate_mse_rmse).reset_index()

# Exibindo os resultados de MSE e RMSE por Canal
print("MSE e RMSE por Canal:")
print(mse_rmse_by_canal)
```

```python
import pandas as pd

# Supondo que forecast_df seja o seu DataFrame inicial com as colunas 'Forecast', 'Saidas', 'Canal', etc.

# Convertendo as colunas para numérico e 'Data FCST' para datetime
forecast_df['Forecast'] = pd.to_numeric(forecast_df['Forecast'], errors='coerce')
forecast_df['Saidas'] = pd.to_numeric(forecast_df['Saidas'], errors='coerce')
forecast_df['Data FCST'] = pd.to_datetime(forecast_df['Data FCST'], errors='coerce')

# Filtrando para valores de 'Saidas' maiores que zero
forecast_df_filtered = forecast_df[forecast_df['Saidas'] > 0].copy()

# Calculando o MSE e RMSE, e transformando o RMSE em uma média percentual
def calculate_mse_rmse_percent(group):
    # Calculando o erro quadrático para cada linha
    group['Erro Quadrático'] = (group['Forecast'] - group['Saidas']) ** 2
    # Calculando o MSE (média do erro quadrático)
    mse = group['Erro Quadrático'].mean()
    # Calculando o RMSE como a raiz quadrada do MSE
    rmse = mse ** 0.5
    # Calculando a média de Saidas para o canal
    media_saidas = group['Saidas'].mean()
    # Transformando o RMSE em percentual em relação à média de Saidas
    rmse_percent = (rmse / media_saidas) * 100
    return pd.Series({'MSE': mse, 'RMSE': rmse, 'RMSE (%)': rmse_percent})

# Aplicando a função por Canal
mse_rmse_percent_by_canal = forecast_df_filtered.groupby('Canal').apply(calculate_mse_rmse_percent).reset_index()

# Exibindo os resultados
print("MSE, RMSE e RMSE (%) por Canal:")
print(mse_rmse_percent_by_canal)
```

```python
file_path = r'C:\Users\fsp_adolpho.salvador\Desktop\Konica Minolta\Desktop Cloud - Documentos\Desktop\Py\Projeto Resultados\Forecast\Arquivos\Forecast.xlsx'
excel_data = pd.ExcelFile(file_path)
forecast_df = excel_data.parse('Forecast')

# Convertendo as colunas 'Forecast' e 'Saidas' para numérico, ignorando erros
forecast_df['Forecast'] = pd.to_numeric(forecast_df['Forecast'], errors='coerce')
forecast_df['Saidas'] = pd.to_numeric(forecast_df['Saidas'], errors='coerce')
forecast_df['Data FCST'] = pd.to_datetime(forecast_df['Data FCST'], errors='coerce')

# Filtrar para incluir apenas os valores onde 'Saidas' > 0
forecast_df_filtered_positive = forecast_df[forecast_df['Saidas'] > 0].copy()

# Adicionar coluna de mês para cálculos mensais
forecast_df_filtered_positive['Mes'] = forecast_df_filtered_positive['Data FCST'].dt.to_period("M")

# Função para calcular o Erro Relativo Normalizado
def calculate_erro_relativo_normalizado(group):
    total_forecast = group['Forecast'].sum()
    total_saidas = group['Saidas'].sum()
    if total_saidas != 0:
        erro_relativo_normalizado = abs(total_forecast - total_saidas) / total_saidas
    else:
        erro_relativo_normalizado = None  # Evita divisão por zero
    return erro_relativo_normalizado

# 1. Erro Relativo Normalizado e Acuracidade por Mês e Canal
erro_relativo_by_month_canal = forecast_df_filtered_positive.groupby(['Canal', 'Mes']).apply(calculate_erro_relativo_normalizado).reset_index(name='Erro Relativo Normalizado')
erro_relativo_by_month_canal['Acuracidade'] = 100 - (erro_relativo_by_month_canal['Erro Relativo Normalizado'] * 100)

# 2. Média de Erro Relativo Normalizado e Acuracidade por Canal
erro_relativo_media_canal = erro_relativo_by_month_canal.groupby('Canal')['Erro Relativo Normalizado'].mean().reset_index(name='Erro Relativo Normalizado Médio')
erro_relativo_media_canal['Acuracidade Média'] = 100 - (erro_relativo_media_canal['Erro Relativo Normalizado Médio'] * 100)

# 3. Erro Relativo Normalizado e Acuracidade por Mês e Material
erro_relativo_by_month_material = forecast_df_filtered_positive.groupby(['Material', 'Mes']).apply(calculate_erro_relativo_normalizado).reset_index(name='Erro Relativo Normalizado')
erro_relativo_by_month_material['Acuracidade'] = 100 - (erro_relativo_by_month_material['Erro Relativo Normalizado'] * 100)

# 4. Média de Erro Relativo Normalizado e Acuracidade por Material
erro_relativo_media_material = erro_relativo_by_month_material.groupby('Material')['Erro Relativo Normalizado'].mean().reset_index(name='Erro Relativo Normalizado Médio')
erro_relativo_media_material['Acuracidade Média'] = 100 - (erro_relativo_media_material['Erro Relativo Normalizado Médio'] * 100)

# Salvando os resultados em um arquivo Excel com todas as abas
output_file_path = 'Erro_Relativo_Normalizado_Acuracidade.xlsx'

with pd.ExcelWriter(output_file_path) as writer:
    erro_relativo_by_month_canal.to_excel(writer, sheet_name='Erro Relativo Mensal', index=False)
    erro_relativo_media_canal.to_excel(writer, sheet_name='Erro Relativo Média por Canal', index=False)
    erro_relativo_by_month_material.to_excel(writer, sheet_name='Erro Relativo por Material', index=False)
    erro_relativo_media_material.to_excel(writer, sheet_name='Erro Relativo Média Material', index=False)

print(f"Arquivo Excel salvo em: {output_file_path}")
```
