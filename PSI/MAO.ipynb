```python
meses = ['04/2024', '05/2024', '06/2024', '07/2024', '08/2024', '09/2024',
         '10/2024', '11/2024', '12/2024', '01/2025', '02/2025', '03/2025']

# Converter colunas numéricas
for col in meses + ["Media_consumo", "Quant estoque 03/2025", "MOS"]:
    mapa_filtrado[col] = pd.to_numeric(mapa_filtrado[col], errors="coerce")

# Calcular o Coeficiente de Variação (CV)
media_consumo_safe = mapa_filtrado["Media_consumo"].replace(0, np.nan)
mapa_filtrado["CV"] = mapa_filtrado[meses].std(axis=1) / media_consumo_safe

# Classificação ABC com base na média de consumo (percentil)
mapa_filtrado["Class_ABC"] = pd.qcut(
    mapa_filtrado["Media_consumo"].fillna(0),
    q=[0, 0.7, 0.9, 1.0],
    labels=["C", "B", "A"]
)

# Classificação XYZ com base em FAIXAS FIXAS de CV
def classificar_xyz(cv):
    if pd.isna(cv):
        return np.nan
    elif cv <= 0.5:
        return "Z"
    elif cv <= 1.0:
        return "Y"
    else:
        return "X"

mapa_filtrado["Class_XYZ"] = mapa_filtrado["CV"].apply(classificar_xyz)

# Combinação final
mapa_filtrado["Class_ABC_XYZ"] = mapa_filtrado["Class_ABC"].astype(str) + mapa_filtrado["Class_XYZ"].astype(str)
```

```python

df = pd.read_json("Mapa_Pareto.json")

# Filtrar apenas as classificações válidas (ignorar 'Cnan' e nulos)
df_valid = df[df["Class_ABC_XYZ"].notna() & (df["Class_ABC_XYZ"] != "Cnan")]

# Contar quantidade por categoria ABC_XYZ
category_counts = df_valid["Class_ABC_XYZ"].value_counts().sort_index()

# Configurar a matriz ABC x XYZ
abc_order = ['A', 'B', 'C']
xyz_order = ['X', 'Y', 'Z']
matrix = pd.DataFrame(0, index=abc_order, columns=xyz_order)

# Preencher a matriz com os dados
for category, count in category_counts.items():
    abc = category[0]
    xyz = category[1]
    if abc in abc_order and xyz in xyz_order:
        matrix.loc[abc, xyz] = count

# Criar o gráfico tipo heatmap
fig, ax = plt.subplots(figsize=(8, 6))
heatmap = ax.imshow(matrix.values, cmap='Blues')

# Adicionar rótulos
ax.set_xticks(range(len(xyz_order)))
ax.set_yticks(range(len(abc_order)))
ax.set_xticklabels(xyz_order)
ax.set_yticklabels(abc_order)
ax.set_xlabel('Classificação XYZ')
ax.set_ylabel('Classificação ABC')
ax.set_title('Matriz ABC/XYZ - Quantidade de Itens')

# Inserir valores dentro das células
for i in range(len(abc_order)):
    for j in range(len(xyz_order)):
        value = matrix.iloc[i, j]
        ax.text(j, i, str(value), ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```
