```python
# 1. Lista com os nomes das colunas de consumo mensal
colunas_meses_12 = ['04/2024', '05/2024', '06/2024', '07/2024', '08/2024', '09/2024',
                    '10/2024', '11/2024', '12/2024', '01/2025', '02/2025', '03/2025']

colunas_meses_6 = ['10/2024', '11/2024', '12/2024', '01/2025', '02/2025', '03/2025']

colunas_meses_3 = ['01/2025', '02/2025', '03/2025']

# 2. Calcular total de consumo nos períodos
total_3 = mapa_kmi[colunas_meses_3].sum().sum()
total_6 = mapa_kmi[colunas_meses_6].sum().sum()
total_12 = mapa_kmi[colunas_meses_12].sum().sum()

# 3. Escolher o conjunto com maior consumo total
if total_12 >= total_6 and total_12 >= total_3:
    colunas_meses = colunas_meses_12
elif total_6 >= total_3:
    colunas_meses = colunas_meses_6
else:
    colunas_meses = colunas_meses_3

# 4. Substituir NaNs por 0 nas colunas de consumo
mapa_kmi[colunas_meses] = mapa_kmi[colunas_meses].fillna(0)

# 5. Calcular média mensal
mapa_kmi['Maior media'] = mapa_kmi[colunas_meses].mean(axis=1)

# 6. Calcular Total IMP (soma das P.O.s futuras)
po_cols = ['P.O. 05/2025', 'P.O. 06/2025', 'P.O. 07/2025', 'P.O. 08/2025', 'P.O. 09/2025']
mapa_kmi['Total IMP'] = mapa_kmi[po_cols].fillna(0).sum(axis=1)

# 7. Calcular o MOS: (Estoque + Total de pedidos) / média de consumo
mapa_kmi['MOS'] = (mapa_kmi['Quant estoque 01/04/2025'] + mapa_kmi['Total IMP']) / mapa_kmi['Maior media'].replace(0, 0.01)
mapa_kmi['MOS'] = mapa_kmi['MOS'].round(2)
```

```python

colunas_meses = ['04/2024', '05/2024', '06/2024', '07/2024', '08/2024', '09/2024',
                 '10/2024', '11/2024', '12/2024', '01/2025', '02/2025', '03/2025']

# 2. Preencher NaN com 0 para não afetar os cálculos
mapa_kmi[colunas_meses] = mapa_kmi[colunas_meses].fillna(0)

# ------------------- ANÁLISE ABC -------------------
df_abc = mapa_kmi[['Material', 'Total Mov.']].copy()
df_abc = df_abc.sort_values(by='Total Mov.', ascending=False)
df_abc['Perc'] = df_abc['Total Mov.'] / df_abc['Total Mov.'].sum()
df_abc['Perc_Acum'] = df_abc['Perc'].cumsum()

def classifica_abc(p):
    if p <= 0.80:
        return 'A'
    elif p <= 0.95:
        return 'B'
    else:
        return 'C'

df_abc['Classe_ABC'] = df_abc['Perc_Acum'].apply(classifica_abc)

# ------------------- ANÁLISE XYZ -------------------
df_xyz = mapa_kmi[['Material'] + colunas_meses].copy()
df_xyz['Média'] = df_xyz[colunas_meses].mean(axis=1)
df_xyz['Desvio'] = df_xyz[colunas_meses].std(axis=1)
df_xyz['CV'] = df_xyz['Desvio'] / df_xyz['Média'].replace(0, 0.01)  # evita divisão por zero

def classifica_xyz(cv):
    if cv <= 0.5:
        return 'X'
    elif cv <= 1.0:
        return 'Y'
    else:
        return 'Z'

df_xyz['Classe_XYZ'] = df_xyz['CV'].apply(classifica_xyz)

# ------------------- JUNTAR AS ANÁLISES -------------------
df_abc_xyz = pd.merge(df_abc[['Material', 'Classe_ABC']], df_xyz[['Material', 'Classe_XYZ']], on='Material')
df_abc_xyz['Classe_ABC_XYZ'] = df_abc_xyz['Classe_ABC'] + df_abc_xyz['Classe_XYZ']
# Primeiro, junta com o resultado da análise ABC
mapa_kmi_2 = mapa_kmi.merge(
    df_abc[['Material', 'Perc', 'Perc_Acum', 'Classe_ABC']],
    on='Material',
    how='left'
)

# Depois, junta com o resultado da análise XYZ
mapa_kmi_2 = mapa_kmi_2.merge(
    df_xyz[['Material', 'Média', 'Desvio', 'CV', 'Classe_XYZ']],
    on='Material',
    how='left'
)

# Por fim, adiciona a coluna combinada ABC_XYZ (opcional)
mapa_kmi_2['Classe_ABC_XYZ'] = mapa_kmi_2['Classe_ABC'] + mapa_kmi_2['Classe_XYZ']

df_abc_xyz.head()
```

```python
df = pd.read_json("mapa_kmi_2.json")

# Filtrar apenas classificações válidas
df_valid = df[df["Classe_ABC_XYZ"].notna() & (df["Classe_ABC_XYZ"] != "Cnan")]

# Contar a quantidade de itens por categoria ABC_XYZ
category_counts = df_valid["Classe_ABC_XYZ"].value_counts().sort_index()

# Definir a ordem das classes
abc_order = ['A', 'B', 'C']
xyz_order = ['X', 'Y', 'Z']
matrix = pd.DataFrame(0, index=abc_order, columns=xyz_order)

# Preencher a matriz com os dados
for category, count in category_counts.items():
    abc = category[0]
    xyz = category[1]
    if abc in abc_order and xyz in xyz_order:
        matrix.loc[abc, xyz] = count

# Criar o heatmap com matplotlib
fig, ax = plt.subplots(figsize=(6, 6))
heatmap = ax.imshow(matrix.values, cmap='Blues')

# Adicionar rótulos dos eixos
ax.set_xticks(range(len(xyz_order)))
ax.set_yticks(range(len(abc_order)))
ax.set_xticklabels(xyz_order)
ax.set_yticklabels(abc_order)
ax.set_xlabel('Classificação XYZ')
ax.set_ylabel('Classificação ABC')
ax.set_title('Matriz ABC/XYZ - Quantidade de Itens')

# Inserir os valores nas células
for i in range(len(abc_order)):
    for j in range(len(xyz_order)):
        valor = matrix.iloc[i, j]
        ax.text(j, i, str(valor), ha='center', va='center', color='black')

plt.tight_layout()
plt.show()
```
